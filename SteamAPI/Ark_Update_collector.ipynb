```python
# Function to parse the raw data and extract details
def parse_forum_data(raw_data):
    """
    This function parses raw forum data to extract update details.

    Parameters:
    - raw_data (str): Raw text data from the forum.

    Returns:
    - df (pd.DataFrame): A DataFrame containing update details.
    """
    # Split the raw data into sections by double newlines
    entries = raw_data.strip().split('\n\n')
    
    # Initialize lists to store extracted data
    dates = []
    versions = []
    num_changes = []
    changes = []
    
    # Regular expression to match the version and date
    version_date_regex = re.compile(r'(v[\d.]+) - (\d+/\d+/\d{2,4})')
    
    for entry in entries:
        # Extract version and date
        version_date_match = version_date_regex.search(entry)
        if version_date_match:
            version = version_date_match.group(1)
            date_str = version_date_match.group(2)
            try:
                date = datetime.strptime(date_str, '%m/%d/%y')
            except ValueError:
                date = datetime.strptime(date_str, '%m/%d/%Y')
            # Extract changes
            entry_lines = entry.strip().split('\n')
            change_list = entry_lines[1:]  # Skip the version/date line
            num_change = len(change_list)
            change_text = ' '.join(change_list)
            # Append to lists
            versions.append(version)
            dates.append(date)
            num_changes.append(num_change)
            changes.append(change_text)
    
    # Create a DataFrame from the parsed data
    df = pd.DataFrame({
        'Date': dates,
        'Version': versions,
        'Num_Changes': num_changes,
        'Changes': changes
    })
    
    return df
```

```python
# Function to categorize updates into minor, major, or before a specific date
def categorize_updates(df):
    """
    Categorizes updates into minor, major, or before a specific date.

    Parameters:
    - df (pd.DataFrame): DataFrame containing update details.

    Returns:
    - df (pd.DataFrame): Updated DataFrame with categories.
    """
    def categorize(version, changes):
        major_keywords = ["major"]
        minor_keywords = ["minor"]
        
        changes_lower = changes.lower()
        if any(keyword in changes_lower for keyword in major_keywords):
            return 'Major'
        elif any(keyword in changes_lower for keyword in minor_keywords):
            return 'Minor'
        else:
            return 'Before'

    df['Category'] = df.apply(lambda row: categorize(row['Version'], row['Changes']), axis=1)
    return df
```
